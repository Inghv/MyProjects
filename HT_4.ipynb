<span style='color:rgb(0,12,102)'><b>9. Сравните среднюю зарплату дата-аналитика и для других топ-5 (по распространенности) профессий (из пункта 5)  
*сравнивайте средние зарплаты только специалистов, соответствующих experience_level!</b></span>

```python
print(ds_salaries.experience_level.value_counts().to_string())
```

```python
SE=ds_salaries.query("(job_title == 'Data Scientist' \
or job_title=='Data Engineer' \
or job_title=='Data Analyst' \
or job_title=='Machine Learning Engineer' \
or job_title=='Research Scientist') and experience_level=='SE'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_SE_in_usd'}).sort_values('avg_payment_of_SE_in_usd',ascending=False)

MI = ds_salaries.query("(job_title == 'Data Scientist' \
or job_title=='Data Engineer' \
or job_title=='Data Analyst' \
or job_title=='Machine Learning Engineer' \
or job_title=='Research Scientist') and experience_level=='MI'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_MI_in_usd'}).sort_values('avg_payment_of_MI_in_usd',ascending=False)

EN = ds_salaries.query("(job_title == 'Data Scientist' \
or job_title=='Data Engineer' \
or job_title=='Data Analyst' \
or job_title=='Machine Learning Engineer' \
or job_title=='Research Scientist') and experience_level=='EN'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_EN_in_usd'}).sort_values('avg_payment_of_EN_in_usd',ascending=False)

EX = ds_salaries.query("(job_title == 'Data Scientist' \
or job_title=='Data Engineer' \
or job_title=='Data Analyst' \
or job_title=='Machine Learning Engineer' \
or job_title=='Research Scientist') and experience_level=='EX'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_EX_in_usd'}).sort_values('avg_payment_of_EX_in_usd',ascending=False)

SE.merge(MI).merge(EN).merge(EX,on='job_title', how='left')
```

<span style='color:rgb(50,200,0)'>Значение NaN в столбцах означает, что Data Scientist'ов ML Engineer'ов и Research Scientist'ов с уровнем EX в таблице нет. Проверим это:</span>

```python
ds_salaries.query("(job_title == 'Data Scientist' \
or job_title=='Machine Learning Engineer' \
or job_title=='Research Scientist') and experience_level=='EX'")
```

<span style='color:rgb(50,200,0)'>Видно, что больше всего зарабатывают Senior'ы Data Scientist'ы, Middle'ы Research Scientist'ы и Junior'ы ML Engineer'ы. Про Expert'ов мало информации.</span>

<span style='color:rgb(0,12,102)'><b>9а. Дополнительно можно сравнить с ‘близкими’ профессиями дата-аналитику, если они не попали в топ-5:</b></span>

<span style='color:rgb(50,200,0)'>Я сравню зарплаты Data Analyst'а с зарплатами Data Science Manager'a,Data Architect'ора,Big Data Engineer'а и Machine Learning Scientist'a. Да, эти специальности, возможно, не слишком близки к Data Analyst'у, но у них по крайней мере размерность выборки (12,11,8 и 8 соответственно) значений позволяет сделать какие-то выводы. У большинства специальностей выборка значений имеет размерность 1 или 2, что недостаточно для того, чтобы сделать какие-то объективные выводы.</span>

```python
SE=ds_salaries.query("(job_title == 'Data Science Manager' \
or job_title=='Data Architect' \
or job_title=='Data Analyst' \
or job_title=='Big Data Engineer' \
or job_title=='Machine Learning Scientist') and experience_level=='SE'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_SE_in_usd'}).sort_values('avg_payment_of_SE_in_usd',ascending=False)

MI = ds_salaries.query("(job_title == 'Data Science Manager' \
or job_title=='Data Architect' \
or job_title=='Data Analyst' \
or job_title=='Big Data Engineer' \
or job_title=='Machine Learning Scientist') and experience_level=='MI'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_MI_in_usd'}).sort_values('avg_payment_of_MI_in_usd',ascending=False)

EN = ds_salaries.query("(job_title == 'Data Science Manager' \
or job_title=='Data Architect' \
or job_title=='Data Analyst' \
or job_title=='Big Data Engineer' \
or job_title=='Machine Learning Scientist') and experience_level=='EN'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_EN_in_usd'}).sort_values('avg_payment_of_EN_in_usd',ascending=False)

EX = ds_salaries.query("(job_title == 'Data Science Manager' \
or job_title=='Data Architect' \
or job_title=='Data Analyst' \
or job_title=='Big Data Engineer' \
or job_title=='Machine Learning Scientist') and experience_level=='EX'") \
.groupby('job_title',as_index=False) \
.agg({'salary_in_usd':'mean'}) \
.rename(columns={'salary_in_usd':'avg_payment_of_EX_in_usd'}).sort_values('avg_payment_of_EX_in_usd',ascending=False)
SE.merge(MI).merge(EN,on='job_title', how='left').merge(EX,on='job_title', how='left')
```

<span style='color:rgb(50,200,0)'>Из полученных результатов заметно, что есть и целые значение и NaN. Целые значения все большие (относительно других). Это свидетельствует о том, что соответствующих записей в исходной таблице мало (1 или 2). Значения NaN свидетельствуют о том, что соответствующих записей в исходной таблице нет.</span>  

<span style='color:rgb(50,200,0)'>Из задания №5 можно заметить, что лучше всего сравнивать средний зарплаты у Data Analyst'ов, Data Scientist'ов,Data Engineer'ов, Machine Learning Engineer'ов и Research Scientist'ов. Результаты, полученные в 9 и 9а лишь подтверждают это. Из полученных результатов в 9 и 9а можно сказать, что самая высокая ЗП у ML Engineer'а из Junior'ов, Research Scientist'а из Middle'ов и Data Scientist'а из Senior'ов. Про Expert'ов сказать ничего нельзя, т.к. некоторые значения в таблице отсутствуют, а те, что нет, получены из 1-2 записей исходной таблицы.</span>
