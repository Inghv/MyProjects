<span style='color:rgb(0,12,102)'><b>1. Загрузить функцию, которая выводит информацию о датасете, переименовать столбцы.</b></span>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
def handle_dataset(file):
    data_frame = pd.read_excel(file)
    data_frame = data_frame.rename(columns=lambda c: c.lower().replace('-', '_').replace('/', '_').replace(' ', '_'))
    print("Информация о таблице: ")
    print("Размерность таблицы (число строк, число столбцов): ")
    print(data_frame.shape)
    print()
    print("Названия столбцов: ")
    print(data_frame.columns)
    print()
    print("Типы данных в столбцах: ")
    print(data_frame.dtypes)
    return data_frame

df = handle_dataset('HT6.xlsx')
df
```

<span style='color:rgb(0,12,102)'><b>2. Создать дополнительные столбцы с ценой и размером прибыли на единицу продукции.</b></span>

```python
df['cost_per_one'] = df.sales/df.quantity
df['profit_per_one'] = df.profit/df.quantity
df
```

<span style='color:rgb(184,41,42)'><b>Построение графиков:</b></span>

<span style='color:rgb(0,12,102)'><b>1. Построить график распредения quantity.</b></span>

```python
plt.figure(figsize=(8, 5))
df.quantity.hist(color=(184/255,41/255,40/255),bins=14)
plt.title('График распределения quantity')
plt.show()
```

<span style='color:rgb(50,200,0)'> *важное замечание: поскольку для построения графика используется встроенная в библиотеку pandas функция, то необходимо изменить параметр bins, отвечающий за количество столбиков гистограммы. Значение этого параметра по умолчанию равно 10.  </span>

<span style='color:rgb(0,12,102)'><b>2. Построить barchart: значение суммарного sales в разрезе регионов.</b></span>

```python
sales_per_region = pd.pivot_table(df,
               index=["region"],
               values=["sales"],
               aggfunc=sum).sort_values('sales',ascending=False)
sales_per_region.plot.bar(color=(255/255,211/255,0/255),grid=True)
```

<span style='color:rgb(0,12,102)'><b>3. Построить barchart: значение суммарного profit в разрезе Sub_category.</b></span>

```python
profit_per_sc = pd.pivot_table(df,
               index=["sub_category"],
               values=["profit"],
               aggfunc=sum).sort_values('profit',ascending=False)
profit_per_sc.plot.bar(color=(51/255,195/255,166/255),grid=True)
```

<span style='color:rgb(0,12,102)'><b>4. Создать датасет, содержащий информацию по стульям. Построить boxplot по регионам, на котором будут отражаться распределения цен в разрезе каждого региона, написать выводы.</b></span>

```python
chairs = df.query("sub_category == 'Chairs'")
chairs.boxplot(column=['cost_per_one'], by=['region'], figsize = (8, 5))
```

<span style='color:rgb(50,200,0)'>Из графиков видно, что минимальная цена на стул в каждом регионе приблизительно одинаковая, а вот максимальная в южном и западном выше, чем в центральном и восточном, примерно на 70 и 50 долларов соответственно. В каждом регионе есть аномальные значения (отмечены точками). Медиана цен в каждом регионе приблизительно одинаковая (в районе 110 долларов).</span>

<span style='color:rgb(0,12,102)'><b>5. Построить график динамики продаж (line chart).</b></span>

```python
sales_per_date = pd.pivot_table(chairs,
               index=["ship_date"],
               values=["sales"],
               aggfunc=sum).sort_values('sales',ascending=False)
sales_per_date.plot(color = (89/225,112/255,174/225),figsize = (10,10))
```

<span style='color:rgb(0,12,102)'><b>6. Построить scatter plot зависимости Sales и Profit - сделать выводы.</b></span>

```python
chairs.plot.scatter(x='sales', y='profit',color = (31/255,181/255,102/255))
```

<span style='color:rgb(50,200,0)'>Можно заметить, что слева есть довольно большое скопление точек, что свидетельствует о том, что заказы от 1 до 1000 продаж приносят в основном либо от 0 до 200 прибыли, либо от 0 до 200 убытка.</span>

<span style='color:rgb(0,12,102)'><b>7. Построить pie chart: доля sales в разрезе Category и доля profit в разрезе Category</b></span>

```python
sales_per_category = pd.pivot_table(df,
               index=["category"],
               values=["sales"],
               aggfunc=sum).sort_values('sales',ascending=False)
sales_per_category = sales_per_category.squeeze()
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
fig.suptitle('Доля sales в разрезе category',fontfamily = 'Times New Roman',
                fontstyle = 'oblique',
                fontsize = 20)
sales_per_category.plot.pie(
    colors=[(0/255,134/255,132/225), (93/225,48/255,58/225), (121/225,71/255,132/225)],
    autopct="%.2f",
    fontsize=10,
    figsize=(5, 5),
)
```

```python
profit_per_category = pd.pivot_table(df,
               index=["category"],
               values=["profit"],
               aggfunc=sum).sort_values('profit',ascending=False)
profit_per_category = profit_per_category.squeeze()
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
fig.suptitle('Доля profit в разрезе category',fontfamily = 'Times New Roman',
                fontstyle = 'oblique',
                fontsize = 20)
profit_per_category.plot.pie(
    colors=[(255/255,141/255,0/225), (47/225,62/255,134/225), (211/225,215/255,13/225)],
    autopct="%.2f",
    fontsize=10,
    figsize=(5, 5),
)
```

<span style='color:rgb(50,200,0)'>Из диаграмм заметно, что доля продаж для каждой категории приблизительно одинакова, а вот наибольшую выручку принесли категории "Технологии" и "Офисные принадлежности".</span>
